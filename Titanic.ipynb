## Filtering Data

```python
#filter based on single condition 
passangers_above_30= data[data["age"]>30]
passangers_above_30
```

```python
# filter based on multiple conditions
# filter male passangers older than 30

male_passangers_above_30= data[(data["age"]>30) & (data["sex"]=="male")]
male_passangers_above_30
```

```python
# filter passangers younger than 18 or in 1st class

young_or_first_class= data[(data["age"] < 18) | (data["pclass"] == 1)]
young_or_first_class
```

```python
# filter passengers who embarked at ports 'C' or 'Q'
embarked_C_Q= data[data["embarked"].isin(['C','Q'])]
embarked_C_Q
```

```python
# filter passangers aged between 20 to 40
age_between_20_to_40= data[data["age"].between(20, 40)]
age_between_20_to_40
```

```python
# filter passangers who are female and survived

female_survived= data.query("sex == 'female' and survived == 1")
print(female_survived)
```

## Handling Missing Data

```python
# handling missing data 
# detect missisng values
data.isnull().sum()
```

```python
# fill missing age with mean age 

data["age"].fillna(data["age"].mean(), inplace= True)
data["age"].isnull().sum()
```

```python
# # Add 'Unknown' to the categories
# data['deck'] = data['deck'].cat.add_categories('Missing')

# Fill missing values with 'Unknown'
data['deck'].fillna('Missing', inplace=True)
```

```python
# drop rows with missing values
data.dropna(subset=['embarked'], inplace=True)
data.isnull().sum()
```

## Data Cleaning and Transformation

```python
# Rename 'Pclass' to 'Passenger Class':
data.rename(columns={'pclass': 'Passenger Class'}, inplace=True)
data.columns
```

```python
# Convert 'Survived' and 'Passenger Class' to categorical
data['Survived'] = data['survived'].astype('category')
data['Passenger_Class'] = data['Passenger Class'].astype('category')
```

```python
# Create a new column 'Family_Size'
data['Family_Size'] = data['sibsp'] + data['parch'] + 1
print(data[['sibsp', 'parch', 'Family_Size']])
```

```python
# Replace 'male' with 'M' and 'female' with 'F'
data['sex'] = data['sex'].replace({'Male': 'M', 'Female': 'F'})
```

## Grouping and Aggregation

```python
# Group by 'Sex' and count the number of survivors
survival_count_by_sex = data.groupby('sex')['survived'].sum()
print(survival_count_by_sex) 
```

```python
# Count non-survivors by sex
non_survivors_count = data[data['survived'] == 0].groupby('sex')['survived'].count()
print(non_survivors_count)
```

```python
# mean age of survivors by 'Pclass'
mean_age_by_class = data.groupby('Passenger Class')['age'].mean()
print(mean_age_by_class)
```

```python
aggregation = data.groupby('Passenger Class').agg({
    'age': ['mean', 'min','max'],
    'fare': 'mean', 
    'survived':'sum'
})
print(aggregation)
```

## Data Visualization

```python
survival = data.groupby('sex')['survived'].sum()
plt.figure(figsize=(5,5))
plt.bar(survival.index, survival.values ,width = 0.5, color=["red", "green"])
plt.title("Survival Count By Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
data['survived'].map({0: 'Did Not Survive', 1: 'Survived'})
sns.countplot(data= data, x="pclass", hue="survived", palette= "spring")
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(data=data, x="survived", y="fare", palette= "GnBu")
plt.title("Fare distribution by Survival")
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.show()
```

```python
plt.hist(data['age'].dropna(), bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
